```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)

school_data.head()
```

```python
# Read Student Data File and store into Pandas DataFrames
student_data = pd.read_csv(student_data_to_load)

student_data.head()
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of schools
total_schools = school_data_complete['school_name'].nunique()
total_schools
```

```python
# Calculate the total number of students
total_students = school_data_complete['Student ID'].nunique()
total_students
```

```python
# Calculate the total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
# Calculate the average math score
avg_math_score = school_data_complete['math_score'].mean()
avg_math_score
```

```python
# Calculate the average reading score
avg_reading_score = school_data_complete['reading_score'].mean()
avg_reading_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
df_passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
percent_passing_math = (df_passing_math['Student ID'].count() / total_students) * 100
percent_passing_math
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
df_passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
percent_passing_reading = (df_passing_reading['Student ID'].count() / total_students) * 100
percent_passing_reading
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
df_overall_passing = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]
percent_overall_passing = (df_overall_passing['Student ID'].count() / total_students) * 100
percent_overall_passing
```

```python
district_summary_df = pd.DataFrame({
    'Total Schools' : [total_schools],
    'Total Students' : [f"{total_students:,}"],
    'Total Budget' : [f"${total_budget:,.2f}"],
    'Average Math Score' : [f"{avg_math_score:.6f}"],
    'Average Reading Score' : [f"{avg_reading_score:.5f}"],
    '% Passing Math' : [f"{percent_passing_math:.6f}"],
    '% Passing Reading' : [f"{percent_passing_reading:.6f}"],
    '% Overall Passing' : [f"{percent_overall_passing:.6f}"]
})

district_summary_df
```

```python
# Average Math Score
school_avg_math_score = school_comparison_df.mean()['math_score']
school_avg_math_score
```

```python
# Average Reading Score
school_avg_reading_score = school_comparison_df.mean()['reading_score']
school_avg_reading_score
```

```python
school_summary_df = pd.merge(school_type, school_comparison_total_students, how="left", on=["school_name"])
```

```python
school_summary_df.rename(columns = {
    'type' : 'School Type',
    'Student ID' : 'Total Students'
}, inplace = True)
school_summary_df
```

```python
school_summary_df['Total School Budget'] = school_comparison_total_budget
school_summary_df['Per Student Budget'] = per_student_budget
school_summary_df['Average Math Score'] = school_avg_math_score
school_summary_df['Average Reading Score'] = school_avg_reading_score
school_summary_df.sort_values('school_name', inplace = True)
school_summary_df
```

```python
df_passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
df_passing_math
```

```python
# % Passing Math
df_passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
school_passing_math = df_passing_math.groupby('school_name')['Student ID'].count()
school_passing_math
```

```python
school_summary_df['Passed Math'] = school_passing_math
school_summary_df
```

```python
school_summary_df['% Passing Math'] = school_summary_df['Passed Math'] / school_summary_df['Total Students'] * 100
school_summary_df.drop('Passed Math', axis='columns', inplace=True)
school_summary_df
```

```python
# % Passing Reading
df_passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
school_passing_reading = df_passing_reading.groupby('school_name')['Student ID'].count()
school_summary_df['Passed Reading'] = school_passing_reading
school_summary_df['% Passing Reading'] = school_summary_df['Passed Reading'] / school_summary_df['Total Students'] * 100
school_summary_df.drop('Passed Reading', axis='columns', inplace=True)
```

```python
# % Overall Passing
df_overall_passing = school_data_complete.loc[(school_data_complete['math_score'] >= 70) 
                                              & (school_data_complete['reading_score'] >= 70)]
school_overall_passing = df_overall_passing.groupby('school_name')['Student ID'].count()
school_summary_df['Overall Passing'] = school_overall_passing
school_summary_df['% Overall Passing'] = school_summary_df['Overall Passing'] / school_summary_df['Total Students'] * 100
school_summary_df.drop('Overall Passing', axis='columns', inplace=True)
```

```python
# Formatting Columns
school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map("${:,.2f}".format)
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map("${:.2f}".format)
```
